## 4. Сколько у каждого из пользователей в среднем покупок в неделю (по месяцам)?


### Определим количество дней в каждом месяце

```python
df_days_in_month = pd.DataFrame(df_full_orders.order_purchase_timestamp.dt.days_in_month).rename(columns={'order_purchase_timestamp':'days_in_month'})
df_days_in_month
```

```python
days_in_date_month = pd.concat([df_full_orders, df_days_in_month], axis=1)[['order_purchase_timestamp', 'days_in_month']]
days_in_date_month
```

### Посчитаем количество покупок каждого пользователя по месяцам

```python
res = df_full_orders\
    .query('order_status == "delivered"')\
    .groupby(['customer_unique_id', 'order_purchase_timestamp'], as_index=False)\
    .agg({'order_id':'nunique'})\
    .rename(columns={'order_id':'cnt_purchase_month'})\
    .sort_values('cnt_purchase_month', ascending=False)
res
```

### Добавим колонку с количеством дней в каждом месяце

```python
res = res\
    .merge(days_in_date_month, on='order_purchase_timestamp', how='left')\
    .drop_duplicates()
res
```

### Посчитаем количество всех покупок пользоватей, общее количество дней в месяцы, когда была совершена покупка и количество таких месяцев

```python
res = res\
    .groupby('customer_unique_id', as_index=False)\
    .agg({'cnt_purchase_month':'sum', 'days_in_month':'sum', 'order_purchase_timestamp':'count'})\
    .rename(columns={'cnt_purchase_month':'cnt_purchases', 'days_in_month':'cnt_days_in_purchases_month', 'order_purchase_timestamp':'cnt_month'})
res    
```

### Расчитаем среднее количество покупок в неделю по пользователям

```python
res['mean_by_weeks'] = res.cnt_purchases / (res.cnt_days_in_purchases_month/res.cnt_month/7)
res.sort_values('mean_by_weeks')
```

## Разделим наших пользовотелей на когорты по первому году и месяцу покупки

### Получим когорты

```python
df_first_purchase = df_full_orders.query('order_status == "delivered"')\
    .groupby('customer_unique_id', as_index=False)\
    .agg({'order_purchase_timestamp': 'min'})\
    .rename(columns={'order_purchase_timestamp': 'first_purchase_date'})

df_first_purchase['first_purchase_date'] = df_first_purchase['first_purchase_date'].dt.to_period('M')

df_first_purchase
```

```python
df_full_orders_cohort = df_full_orders.merge(df_first_purchase, on='customer_unique_id', how='left')
df_full_orders_cohort.head()
```

```python
df_full_orders_cohort['purchase_month'] = df_full_orders['order_purchase_timestamp'].dt.to_period('M')
df_full_orders_cohort.head()
```

```python
cohort_data = df_full_orders_cohort\
    .query('order_status == "delivered"')\
    .groupby(['first_purchase_date', 'purchase_month'])\
    .agg({'customer_unique_id': 'nunique'})\
    .rename(columns={'customer_unique_id': 'active_users'})\

cohort_data
```

```python
cohort_data = cohort_data.reset_index()
```

```python
cohort_pivot = cohort_data.pivot(index='first_purchase_date', columns='purchase_month', values='active_users').fillna(0)
cohort_pivot
```

```python
cohort_pivot = cohort_data.pivot(index='first_purchase_date', columns='purchase_month', values='active_users')

# Plot the heatmap
plt.figure(figsize=(14, 7))
sns.heatmap(cohort_pivot, annot=True, fmt=".0f", cmap="YlGnBu")
plt.title('Когортный анализ - Активные пользователи по месяцам')
plt.xlabel('Месяц покупки')
plt.ylabel('Когорта (Месяц первой покупки)')
plt.show()
```

1) Начало каждой когорты:<br><br>  
    В первый месяц после первой покупки количество активных пользователей наибольшее. Это ожидаемо, так как все пользователи в когорте активны в этот период.<br><br>

4) Долгосрочная активность: Некоторые когорты показывают устойчивую активность на протяжении нескольких месяцев, что может свидетельствовать о лояльности пользователей.<br><br>

## 5.2. В период с января по декабрь выявите когорту с самым высоким retention на 3-й месяц.

### Добавим колонку с количетвом месяцов, пройденных с первой покупки

```python
df_full_orders_cohort_delivered = df_full_orders_cohort.query('order_status == "delivered"')
df_full_orders_cohort_delivered.head()
```

```python
df_full_orders_cohort_delivered['cohort_month'] = (df_full_orders_cohort_delivered['purchase_month'] - df_full_orders_cohort_delivered['first_purchase_date']).apply(lambda x: x.n)
df_full_orders_cohort_delivered.head()
```

```python
cohort_counts = df_full_orders_cohort_delivered\
    .groupby(['first_purchase_date', 'cohort_month'])\
    .agg({'customer_unique_id': 'nunique'})\
    .rename(columns={'customer_unique_id': 'active_users'})
cohort_counts
```

### Посчитаем удержание

```python
cohort_sizes = cohort_counts\
    .reset_index()\
    .groupby('first_purchase_date')\
    .agg({'active_users':'sum'})\
    .rename(columns={'active_users': 'cohort_size'})

cohort_sizes
```

```python
retention = cohort_counts.unstack(level=1).divide(cohort_sizes['cohort_size'], axis=0)
retention
```

```python
third_month_retention = retention[('active_users', 3)]
third_month_retention
```

```python
max_retention_cohort = third_month_retention.idxmax()
max_retention_value = third_month_retention.max()

print(f"Когорта с самым высоким удержанием на 3-й месяц: {max_retention_cohort}, Удержание: {max_retention_value:.2%}")
```

## 6. Часто для качественного анализа аудитории использую подходы, основанные на сегментации. Используя python, построй RFM-сегментацию пользователей, чтобы качественно оценить свою аудиторию. Для каждого RFM-сегмента построй границы метрик recency, frequency и monetary для интерпретации этих кластеров.

### Актуальная дата

```python
import datetime as dt
current_date = dt.datetime.today()
current_date
```

### Найдём Recency (количество времени, прошедшего с момента первой покупки)

```python
df_full_orders_delivered= df_full_orders.query('order_status == "delivered"')
```

```python
recency = df_full_orders_delivered.groupby('customer_unique_id')['order_purchase_timestamp'].max().reset_index()
recency['recency'] = (current_date - recency['order_purchase_timestamp']).dt.days
recency
```

### Найдём Frequency (суммарное количество покупок у пользователя за всё время)

```python
frequency = df_full_orders_delivered.groupby('customer_unique_id').size().reset_index(name='frequency')
frequency
```

### Найдём Monetary (общая сумма покупок)

```python
monetary = df_full_orders_delivered.groupby('customer_unique_id')['price'].sum().reset_index(name='monetary')
monetary
```

### Объединим метрики

```python
rfm = recency.merge(frequency, on='customer_unique_id').merge(monetary, on='customer_unique_id')
rfm
```

 ## Создание RFM-сегментов

```python
quintiles = rfm[['recency', 'frequency', 'monetary']].quantile([.2, .4, .6, .8]).to_dict()
quintiles
```

```python
def r_score(x):
    if x <= quintiles['recency'][.2]:
        return 5
    elif x <= quintiles['recency'][.4]:
        return 4
    elif x <= quintiles['recency'][.6]:
        return 3
    elif x <= quintiles['recency'][.8]:
        return 2
    else:
        return 1

def fm_score(x, c):
    if x <= quintiles[c][.2]:
        return 1
    elif x <= quintiles[c][.4]:
        return 2
    elif x <= quintiles[c][.6]:
        return 3
    elif x <= quintiles[c][.8]:
        return 4
    else:
        return 5   
```

```python
rfm['R'] = rfm['recency'].apply(lambda x: r_score(x))
rfm['F'] = rfm['frequency'].apply(lambda x: fm_score(x, 'frequency'))
rfm['M'] = rfm['monetary'].apply(lambda x: fm_score(x, 'monetary'))
```

```python
rfm
```

```python
rfm['RFM_Score'] = rfm['R'].map(str) + rfm['F'].map(str) + rfm['M'].map(str)
rfm.head()
```

```python
rfm_summary = rfm.groupby('RFM_Score').agg({
   'recency': ['min', 'max'],
   'frequency': ['min', 'max'],
   'monetary': ['min', 'max']
}).reset_index()
```

```python
rfm_summary
```

```python
for index, row in rfm_summary.iterrows():
       print(f"\n RFM-сегмент {row['RFM_Score'][0]} имеет границы:")
       print(f"  Recency: от {row[('recency', 'min')]} до {row[('recency', 'max')]} дней")
       print(f"  Frequency: от {row[('frequency', 'min')]} до {row[('frequency', 'max')]} заказов")
       print(f"  Monetary: от {row[('monetary', 'min')]} до {row[('monetary', 'max')]} рублей")
```

### Разделение пользователей на сегменты

```python
segt_map = {
    r'[1-2][1-2]': 'Hibernating',
    r'[1-2][3-4]': 'At risk',
    r'[1-2]5': "Can't lose",
    r'3[1-2]': 'About to sleep',
    r'33': 'Need attention',
    r'[3-4][4-5]': 'Loyal customers',
    r'41': 'Promising',
    r'51': 'New customers',
    r'[4-5][2-3]': 'Potential loyalists',
     r'5[4-5]': 'Champions'
}


rfm['Segment'] = rfm['R'].map(str) + rfm['F'].map(str)
rfm['Segment'] = rfm['Segment'].replace(segt_map, regex=True)


rfm.head()
```

### Сегменты:

<h3>
<ul>
    <li>Hibernating: Пользователи с низкой частотой и давней последней покупкой.</li><br><br>
    <li>At risk: Пользователи, которые покупали часто, но давно.</li><br><br>
    <li>Can't lose: Очень ценные пользователи, которые давно не покупали.</li><br><br>
    <li>About to sleep: Пользователи, которые могут перестать быть активными.</li><br><br>
    <li>Need attention: Пользователи, которые требуют дополнительного внимания для удержания.</li><br><br>
    <li>Loyal customers:Часто покупающие пользователи.</li><br><br>
    <li>At risk: Пользователи, которые покупали часто, но давно.</li><br><br>
    <li>Potential loyalists: Пользователи, которые могут стать лояльными при правильном подходе.</li><br><br>
    <li>New customers: Недавно привлеченные пользователи.</li><br><br>
    <li>Champions: Лучшие клиенты с высокой активностью и суммой покупок.</li><br><br>
</ul>
</h3>

```python
ax = sns.countplot(data=rfm, y='Segment')
ax.set_title('График количества клиентов по сегментам')
ax.set_xlabel('Количество клиентов')
ax.set_ylabel('Сегменты')
```

### Краткий анализ:
#### 1) Hibernating: Самый большой сегмент. Эти клиенты давно не совершали покупок и имеют низкую частоту. Возможно, стоит рассмотреть стратегии повторного вовлечения.
#### 2) New customers: Значительное количество новых клиентов. Важно удержать их и превратить в лояльных.
#### 3) About to sleep: Клиенты, которые могут перестать быть активными. Необходимо стимулировать их к покупкам.
#### 4) Promising: Клиенты с потенциалом. Стоит развивать их лояльность.
#### 5) Can't lose: Ценные клиенты, которые давно не покупали. Важно вернуть их активность.
#### 6) Loyal customers: Эти клиенты часто покупают. Необходимо поддерживайть их лояльность через программы поощрения.
#### 7) Champions: Лучшие клиенты. Необходимо продолжайть поддерживать их интерес и удовлетворенность.
