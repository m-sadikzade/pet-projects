## 2. Сколько заказов в месяц в среднем не доставляется по разным причинам (вывести детализацию по причинам)?


```python
df_full_orders.head()
```

### Получим количество месяцев в наших данных

```python
months_cnt = df_full_orders.order_purchase_timestamp.dt.month.nunique()
months_cnt
```

#### Заказы, которые не были доставлены

```python
orders_not_delivered = df_full_orders[df_full_orders['order_delivered_customer_date'].isnull()]\
                       .query('order_status != "delivered"')
orders_not_delivered
```

```python
orders_not_delivered_by_status_cnt = orders_not_delivered\
    .groupby('order_status', as_index=False)\
    .agg({'order_id':'count'})\
    .rename(columns={'order_id':'cnt'})\
    .sort_values('cnt', ascending=False)
orders_not_delivered_by_status_cnt
```

```python
orders_not_delivered_by_status_cnt['avg_cnt_by_month'] = orders_not_delivered_by_status_cnt.cnt / months_cnt
```

```python
orders_not_delivered_by_status_cnt
```

```python
plt.figure(figsize=(15, 5))
ax = sns.barplot(data=orders_not_delivered_by_status_cnt, x='avg_cnt_by_month', y='order_status')
ax.set_title('Колочество отменённых заказов по различным причинам в среднем за месяц')
ax.set_xlabel('Количество')
ax.set_ylabel('Причины')
```

## 3. По каждому товару определить, в какой день недели товар чаще всего покупается.


### Добавим отдельную колонку с днём недели для удобства

```python
df_full_orders['weekday'] = df_full_orders.order_purchase_timestamp.dt.weekday.apply(lambda x: x+1)
```

```python
df_full_orders.head()
```

### Найдём количество покупок для каждого товара по дням недели

```python
purchases_by_product_id_and_weekday = df_full_orders\
    .query('order_status == "delivered"')\
    .groupby(['product_id', 'weekday'], as_index=False)\
    .agg({'order_id':'nunique'}).rename(columns={'order_id':'cnt'})\
    .pivot(index='weekday', columns='product_id', values='cnt')
purchases_by_product_id_and_weekday
```

```python
def get_max_days(series):
    max_value = series.max()
    return series.index[series == max_value].tolist()
```

```python
max_weekdays_per_product = purchases_by_product_id_and_weekday.apply(get_max_days)
```

```python
max_weekdays_per_product = max_weekdays_per_product.reset_index().rename(columns={0: 'max_purchases_by_weekdays'})
```

```python
max_weekdays_per_product
```

## 4. Сколько у каждого из пользователей в среднем покупок в неделю (по месяцам)?


### Определим количество дней в каждом месяце

```python
df_days_in_month = pd.DataFrame(df_full_orders.order_purchase_timestamp.dt.days_in_month).rename(columns={'order_purchase_timestamp':'days_in_month'})
df_days_in_month
```

### Добавим полученные данные в основную таблицу

```python
df_full_orders = pd.concat([df_full_orders, df_days_in_month], axis=1)
df_full_orders.head()
```

### Расчитаем среднее количество покупок в неделю по пользователям

```python
purchases_by_customer = df_full_orders[['customer_id', 'product_id', 'order_status', 'days_in_month']]
purchases_by_customer = purchases_by_customer\
    .query('order_status == "delivered"')\
    .groupby(['customer_id', 'days_in_month'], as_index=False)\
    .agg({'product_id':'nunique'})\
    .rename(columns={'product_id':'cnt'})\
    .sort_values('cnt', ascending=False)

purchases_by_customer    
```

```python
purchases_by_customer['avg_weekly_purchases'] \
= purchases_by_customer.cnt / (purchases_by_customer.days_in_month/ 7)

purchases_by_customer
```
